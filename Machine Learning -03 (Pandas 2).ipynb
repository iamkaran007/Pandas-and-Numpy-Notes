## Reorganize Data

```python
weather = pd.read_csv('data/weather.csv')
weather.head()
```

```python
weather_melt= pd.melt(weather,
        id_vars=['id','year','month','element'],
       var_name='day',
        value_name='temp'
       ) #to make them in 1 column, id_vars are those, whom we don't want to melt
weather_melt.head()
```

```python
#now we want to do other operation
weather_tidy = weather_melt.pivot_table(index=['id','year','month','day'], #they are those that are not gonna change
                        columns='element',
                        values='temp',dropna=False).reset_index()
weather_tidy.head()
```

```python
weather_tidy.to_csv('output/weather.csv')
```

```python
# whats the max temp on 15 Jan
weather_tidy.loc[(weather_tidy.month ==1) &(weather_tidy.day == 'd15'),['tmax']]
```

```python
def avg(arr):
    return sum(arr)/len(arr)
```

```python
weather_tidy[['month','year']].apply(avg)
```

```python
weather_tidy[['month','year']].apply(avg,axis=1).head() # gives average of month and year
```

```python
def extract(arr):
    return [int(dayStr[1:]) for dayStr in arr]  # we have to pass column not only string
#extract('day12')

def extract2(arr):
    dayStr= arr[0]
    return [int(dayStr[1:])]
```

```python
day= weather_tidy[['day']].apply(extract)
day.head()
```
