### Operations  on ROWS

```python
#loc --> work on labels and iloc --> work on index as when we delete the row_label 3 , then index will get updated and we will get 1 2 4 5 for labels and 1 2 3 4 for index
```

```python
df.loc[100] # works as dictionary
```

```python
df.loc[[10,100]]
```

```python
df.iloc[[10,100]] # as of now labels and index are same
```

```python
df.iloc[-1]
```

```python
#df.loc[-1] --> error
```

```python
df.iloc[3:7] # here 7 is not included
```

```python
df.loc[3:7] # here 7 is inculded
```

```python
df.iloc[3:10,1:4] # pehle rows and then columns sams as numpy
```

```python
#df.loc[3:7,1:4] --> error as they are labels not indexes

df.loc[3:7,['country','year']]
```

```python
df.iloc[[1,10,100],[0,2,3]] # instead of range, we can also pass them in list
```

```python
m = df['lifeExp'].mean() # It is a Series
m
```

```python
a = df['lifeExp'] > 59
a.head()
```

```python
#here we are working on rows so, we will use loc and iloc
df.loc[df.lifeExp >59].head() # or df.loc[df['lifeExp'] >59] first one wala is a shortcut
```

```python
df.loc[df.lifeExp >m].head()
```

```python
df.loc[df.lifeExp >59,['lifeExp','country']].head() # we pass names here as we have use loc
```

```python
# select rows where lifeExp is between 51 and 53
```

```python
df.loc[(df.lifeExp > 51) & (df.lifeExp <53)].head() # keep them in brackets
```

```python
# rows where lifeexp <20 or >65
df.loc[(df.lifeExp <25) | (df.lifeExp >60)].head()
```

```python
# select rows where country is India

df.loc[df.country =='India']
```

```python
df.loc[df.country >'India'].head() # Afer India ,means from indonesia to zimbavwe
```
